## Elasticity as a Concept 

Let's say that you are starting a new business. When you start a new business, you want to figure out how you want to price your products so that you can get the most people to buy them and at the same time, make a lot of money. How the fuck do we do this? 

Let's start with a simple product. 

You are selling ladoos in a sweet shop. You decide that your ladoos are the best in the entire city! Because of this, you want to make your ladoos more expensive than everyone else's. Let's say that the average price of a ladoo at some random shop is 8 rupees. You decide that you want to sell your ladoos for 20 rupees. If you decide to sell your ladoos at 20 rupees, how does this affect the quantity demanded by customers? 

$$ Elasticity_D = \frac{\% \bigtriangleup Q}{\% \bigtriangleup P} $$ 

In order to look at elasticity, let's first create a demand graph for an average ladoo shop. 

```python
x_ladoo = np.array([10,20,30,40,50]) 
y_ladoo = np.array([24,16,8,4,2])  
fig, ax = plt.subplots(figsize=(14,8))
ax.plot(x_ladoo, y_ladoo, 'b-')
plt.axhline(y=8, xmin=0.16, xmax=0.5, color='g')
plt.axvline(x=15, ymin=0.301, ymax=0.79) 
ax.text(x=9, y=14, s=r"$\bigtriangleup P = +12$", fontsize=15)
ax.text(x=16, y=14, s=r"$\uparrow $", fontsize=30)
ax.text(x=19, y=9, s=r"$\bigtriangleup Q = -15 $", fontsize=15)
ax.text(x=20, y=7, s=r"$\leftarrow $", fontsize=30)
ax.set_title("Ladoo Shop Demand", size=24)
ax.set_xlabel("Quantity", size=20)
ax.set_ylabel("Price", size=20) 
```

### Change in Price 

To get the change in price, we start with the price we want to move from. In this case, we want to move from 8 rupees to 20 rupees. 

So the change in price can be shown by a math symbol known as delta which is basically a triangle. 

Delta looks like this $$ \bigtriangleup $$ 

So the change in price or the delta of the price is calculated like this: 

$$ \bigtriangleup P = P_{new} - P_{old} = 20 - 8 = 12 $$ 

$$ P_{old} = 8 $$

$$ P_{new} = 20 $$ 

So the change in quantity or the delta of the quantity is calculated like this:

$$ \bigtriangleup Q = Q_{new} - Q_{old} = 15 - 30 = -15 $$ 

$$ Q_{old} = 30 $$

$$ Q_{new} = 15 $$  

Calculating Elasticity of Demand 

We first need to get the average of the two different prices and quantities. 

Average of Prices

$$ \mu_{P} = \frac{P_{new} + P_{old}}{2} $$ 

Average of Quantities 

$$ \mu_{Q} = \frac{Q_{new} + Q_{old}}{2} $$ 

Let's calculate the averages of our ladoo shop problem. 

$$ P_{old} = 8 $$

$$ P_{new} = 20 $$

$$ \mu_{P} = \frac{8 + 20}{2} = \frac{28}{2} = 14 $$

$$ Q_{old} = 30 $$

$$ Q_{new} = 15 $$

$$ \mu_{Q} = \frac{30 + 15}{2} = \frac{45}{2} = 22.5 $$

Calculating Percentages 

Percentage of Price

$$ \frac{\bigtriangleup P}{\mu_{P}} * 100 = \frac{12}{14} * 100 = \frac{6}{7} * 100 = 85.7 \% $$ 

```python
6/7 * 100
```

Percentage of Quantity 

$$ \frac{\bigtriangleup Q}{\mu_Q} * 100 = \frac{-15}{22.5} * 100 = \frac{-3}{4.5} * 100 = -66.67 \% $$ 

```python
-3 / 4.5 * 100 
```

$$ E_D = \frac{-66.67}{85.7} = -0.7779 $$ 

```python
-66.67 / 85.7
```

Calculating Elasticity Part II 

Cross Multiplication Method 

$$ E_D = \frac{\frac{\bigtriangleup Q}{\mu_Q} * 100}{\frac{\bigtriangleup P}{\mu_P} * 100} $$ 

$$ E_D = \frac{\frac{-15}{22.5} * 100}{\frac{12}{14} * 100} $$ 

$$ E_D = \frac{\frac{-15}{22.5}}{\frac{12}{14}} $$ 

$$ E_D = \frac{-15}{22.5} \times \frac{14}{12} $$ 

$$ E_D = \frac{7}{6} \times \frac{-3}{4.5} $$ 

$$ E_D = \frac{7 * -3}{6 * 4.5} = \frac{-21}{27} = \frac{-7}{9} = -0.7779 $$
