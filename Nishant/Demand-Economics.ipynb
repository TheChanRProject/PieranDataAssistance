## Studying Basic Law of Demand 

So what the fuck is demand? 

If two people are selling a gaming account which has all the same stuff in it, but with a different name, would you buy one that is at a higher price or a lower price? 

If you are like most people, you would probably buy the one that is priced lower. The reason for this is because of something called the Law of Demand. 

The Law of Demand means that as the price goes up, the less demand there is for a product. 

Let's see this in the form of an equation which has a graph shown below where P is a function of the quantity that is demanded.

$$ P = f(Q) $$ 

```python
Quantity = [10,20,30,40,50]
Price = [100,80,60,40,20] 

xQ = np.array(Quantity)
yP = np.array(Price)
fig, ax = plt.subplots(figsize=(14,8)) 
plt.plot(xQ,yP,'r') 
ax.set_title("Law of Demand", size=24)
ax.set_xlabel("Quantity", size=20)
ax.set_ylabel("Price", size=20)

```
